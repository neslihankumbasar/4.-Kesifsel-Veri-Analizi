```python
### Soru 1 :
## Toplam gelir verilerini içeren TOTAL_REVENUE ve toplam harcama verilerini içeren TOTAL_EXPENDITURE değişkenlerini inceleyin. 
## Bu değişkenler aykırı değerler içeriyor mu?
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

states=pd.read_csv("C:/Users/neslihan.kumbasar/bootcamp/states_all.csv")
```

```python
plt.figure(figsize=(15,5))

baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.subplot(1,2,1)
plt.hist(states['TOTAL_REVENUE'] )
plt.title("TOTAL_REVENUE", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.hist(states['TOTAL_EXPENDITURE'] )
plt.title("TOTAL_EXPENDITURE", fontdict=baslik_font)

plt.show()

### Grafiklerden de gordugumuz uzere iki degiskende de aykiri degerlerin oldugunu rahatlikla gozlemliyebiliyoruz.
```

```python
### Soru 2 :
## Toplam gelir (TOTAL_REVENUE) ve toplam harcama (TOTAL_EXPENDITURE) değişkenlerinde aykırı değer tespit ettiyseniz 
## bu derste öğrendiğiniz teknikleri uygulayarak bu aykırı değerleri ortadan kaldırın ve 
## bunu yaptıktan sonra hiçbir aykırı değer kalmadığını doğrulayın.
```

```python
plt.figure(figsize=(10,5))

baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }

plt.subplot(2,2,1)
plt.hist(states['TOTAL_REVENUE'] )
plt.title("TOTAL_REVENUE", fontdict=baslik_font)

plt.subplot(2,2,2)
plt.hist(np.log(states['TOTAL_REVENUE']) )
plt.title("TOTAL_REVENUE (log donusumu yapilmis)", fontdict=baslik_font)

plt.subplot(2,2,3)
plt.hist(states['TOTAL_EXPENDITURE'] )
plt.title("TOTAL_EXPENDITURE", fontdict=baslik_font)

plt.subplot(2,2,4)
plt.hist(np.log(states['TOTAL_EXPENDITURE']) )
plt.title("TOTAL_EXPENDITURE (log donusumu yapilmis)", fontdict=baslik_font)

plt.tight_layout()
plt.show()

### Aykiri degerler log donusumu sayesinde normal dagilima donustu.
```

```python
### Soru 3 :
## Toplam gelir (TOTAL_REVENUE) değişkeninden, toplam harcama (TOTAL_EXPENDITURE) değerini çıkartarak bütçe açığı olarak 
## düşünülebilecek bir değişken oluşturun (aykırı değerleri temezlemeden). 
## Bu yeni değişkende de aykırı değerler yer almakta mı? Varsa bu aykırı değerleri de temizleyin.
```

```python
TOTAL_REVENUE= states['TOTAL_REVENUE']
TOTAL_EXPENDITURE= states['TOTAL_EXPENDITURE']
butce_acigi= TOTAL_REVENUE - TOTAL_EXPENDITURE
butce_acigi

plt.figure(figsize=(15,5))

baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }

plt.subplot(1,2,1)
plt.hist(butce_acigi)
plt.title("Butce Acigi", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.hist(np.log(butce_acigi) )
plt.title("Butce Acigi (log donusumlu)", fontdict=baslik_font)

plt.show()
```

```python
### Soru 4 :
## Yine toplam gelir (TOTAL_REVENUE) değişkeninden, toplam harcama(TOTAL_EXPENDITURE) değerini çıkartarak bütçe açığı olarak 
## düşünülebilecek bir değişken oluşturun. 
## Fakat bu sefer aykırı değerleri temezledikten sonra bu işlemi yapın. 
## Bu yeni değişkende de aykırı değerler var mı? Varsa bunları da temizlemekte fayda var.
```

```python
total_revenue_log= np.log(states['TOTAL_REVENUE'])
total_expenditure_log= np.log(states['TOTAL_EXPENDITURE'])
butce_acigi=total_revenue_log-total_expenditure_log

plt.figure(figsize=(10,5))

baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }

plt.subplot(1,2,1)
plt.hist(butce_acigi)
plt.title("Butce Acigi", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.hist(np.log(butce_acigi))
plt.title("Butce Acigi (log donusumlu)", fontdict=baslik_font)

plt.tight_layout()

plt.show()


## Hesaplama oncesinde degiskenlerde ayri ayri loglama islemi yaptigimiz halde yine 
## elimizdeki yeni degiskende aykiri degerlerin oldugunu gozlemliyoruz.Bu yuzden loglama islmeini tekrarliyoruz.
## cikan sonucta bir miktar daha normale dondugunu goruyoruz.


```
